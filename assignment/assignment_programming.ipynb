### **Q2.** **Being able to make basic plots to visualize sets of points is really helpful, particularly with data analysis. The pyplot plots are built up slowly by defining elements of the plot, and then using `plt.show()` to create the final plot.**  



1. Import the `numpy` module as `np` and the `matplotlib.pyplot` package as `plt`.

```python
import numpy as np
import matplotlib.pyplot as plt
```


2. Use `np.linspace` to create a grid of 50 points ranging from 0 to 1.

```python
grid = np.linspace(0, 1, 50)
print(grid)
```

3. Create a numpy array $y$ containing the values for the natural logarithm function using the `np.log(x)` function. Create a numpy array $z$ containing the values for the exponential function using the `np.exp(x)` function.

```python
log_arr = np.log(grid)
print(log_arr)
exp_arr = np.exp(grid)
print(exp_arr)
```


4. Use the `plt.scatter(x,y)` method for the $y$ and $z$ vectors to create two scatter plots of the points you've created.
5. Use the plt.show() method to create the plot.

```python
plt.scatter(grid,log_arr)
plt.scatter(grid,exp_arr)
plt.show()
```

6. Before the plt.show() call, add labels to the and axes using plt.xlabel(label) and plt.ylabel(label). Add a title to the graph using plt.title(title), like "Natural Log and Exponential Functions".

```python
grid2 = np.linspace(0, 1, 50)
log_arr2 = np.log(grid2)
exp_arr2 = np.exp(grid2)
plt.scatter(grid2, log_arr2, label = "Natural Logarithm")
plt.scatter(grid2, exp_arr2, label = "Exponential")
plt.title("Natural Logarithm and Exponential Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

7. That looks a lot better, but we need a legend. When you screate the scatter plots, add the argument label='Natural Log' or label='Exponential' to the .scatter method call. Before the plt.show() call, add a plt.legend(loc = 'lower right') method call, which creates a legend in the lower right.

```python
grid3 = np.linspace(0, 1, 50)
log_arr3 = np.log(grid3)
exp_arr3 = np.exp(grid3)
plt.scatter(grid3, log_arr3, label = "Natural Logarithm")
plt.scatter(grid3, exp_arr3, label = "Exponential")
plt.title("Natural Logarithm and Exponential Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc = "lower right")
plt.show
```
